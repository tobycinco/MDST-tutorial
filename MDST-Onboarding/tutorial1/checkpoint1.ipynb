```python
# @@@ 5
df["INSTRUCT_OVER_TOTAL"] = df["INSTRUCTION_EXPENDITURE"] / df["TOTAL_EXPENDITURE"] 
df.head()
```

**Feature engineering justification:** I added a column "INSTRUCT OVER TOTAL" which is the instruction expenditure divided by the total expenditure because I wanted to see if having a greater percentage of the total expenditure going towards instruction would have an impact on scores.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
# @@@ 8

X = df[['ENROLL_4','AVG_MATH_8_SCORE','AVG_READING_4_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_4_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
# @@@ 9 

# X_train, X_test, y_train, y_test = train_test_split(
#      X, y, test_size=, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# @@@ 10
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# @@@ 11
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```
